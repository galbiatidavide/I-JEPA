```python
from PIL import Image
import numpy as np

# Carica l'immagine
image_path = 'mountains.jpg'
image = Image.open(image_path)

# Ridimensiona l'immagine a 60x100 pixel
resized_image = image.resize((100, 60))

# Converti l'immagine in scala di grigi
gray_image = resized_image.convert('L')

# Converti l'immagine in un array NumPy
image_array = np.array(gray_image)

# Normalizza i valori dei pixel tra 0 e 1
normalized_array = image_array / 255.0

# Salva l'array normalizzato come file .npy
npy_path = 'mountains.npy'
np.save(npy_path, normalized_array)

print(f"File salvato come {npy_path}")
```
